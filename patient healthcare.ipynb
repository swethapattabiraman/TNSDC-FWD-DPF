```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io

# Upload CSV file
print("Please upload the dataset file...")
uploaded = files.upload()

# Get uploaded filename dynamically
file_name = list(uploaded.keys())[0]
print("Uploaded file:", file_name)

# Load dataset
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# Dataset loaded message
print("\nHealthcare Dataset Loaded Successfully")
print("=====================================")

# Display first 5 rows
print("\nFirst 5 Records:")
print(df.head())

# Dataset dimensions
rows, columns = df.shape
print("\nTotal Number of Rows:", rows)
print("Total Number of Columns:", columns)

# Display column names
print("\nColumn Names in Dataset:")
for col in df.columns:
    print("-", col)

# Check missing values
print("\nMissing Values Check:")
print(df.isnull().sum())

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Disease count
disease_count = df["Disease"].value_counts()
top_5_diseases = disease_count.head(5)

print("\nTop 5 Diseases and Patient Count:")
print(top_5_diseases)

# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(top_5_diseases.index, top_5_diseases.values)
plt.xlabel("Disease Name")
plt.ylabel("Number of Patients")
plt.title("Healthcare Patient Monitoring - Bar Chart")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(
    top_5_diseases.values,
    labels=top_5_diseases.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Healthcare Patient Distribution - Pie Chart")
plt.axis("equal")
plt.show()

# Completion message
print("\nBar Chart and Pie Chart Displayed Successfully")
print("Project Execution Completed Successfully")
```
